```python
# Goal:
# Compare SBPBefore vs SBPAfter for the SAME subjects
# using a paired (dependent) t-test. (SBPBefore = Systolic blood pressure before taking a drug, SBPAfter = Systolic blood pressure after taking a drug).
```

```python
# Step 1: Import required libraries
#
# pandas -> read Excel file
# scipy.stats -> paired t-test
# google.colab.files -> upload file in Google Colab

import pandas as pd
from scipy import stats
from google.colab import files
```

```python
# Cell 2 — Upload the Excel input file
# Choose: paired_ttest_pressure_data.xlsx

uploaded = files.upload()
input_filename = list(uploaded.keys())[0]
print("Uploaded file:", input_filename)
```

```python
# Step 3: Read the Excel file

data = pd.read_excel(input_filename)

# Show the first few rows
data
```

```python
# Step 4: Extract paired columns
#
# These are the same subjects measured twice

sbp_before = data["SBPBefore"]
sbp_after  = data["SBPAfter"]
```

```python
# Step 5: Perform paired t-test
#
# H0: Mean(SBPBefore - SBPAfter) = 0
# H1: Mean(SBPBefore - SBPAfter) != 0

t_stat, p_value = stats.ttest_rel(sbp_before, sbp_after)

print("Paired t-test results")
print("t statistic:", round(t_stat, 4))
print("p-value:", round(p_value, 4))
```
